```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('appacc22232016-07-11_ipm_data.txt')
max=np.amax(data)
ind = np.where(max==data)
print(ind)
```

```python
plt.plot(data)
ax = plt.axes()
ax.set_facecolor("black")
plt.plot(ind,max,'.',markersize=7,color ='r')
plt.grid()
plt.savefig("Ex3_plot.png")
```
